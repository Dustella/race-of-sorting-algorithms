```python
import pandas as pd
import matplotlib.pyplot as plt

result = pd.read_csv("./timing.csv")
result
```

```python
# draw a barh plot
# sname,10_time,100_time,1000_time,10000_time,100000_time
# BubbleSort,38,414,41708,4714321,480703471
# HeapSort,24,151,1903,32661,322298
# HillSort,45,123,1531,24756,335565
# InsertSort,22,128,10369,1092291,109279925
# MergeSort,97,749,7124,84289,938115
# QuickSort,29,115,894,11136,133703
# SelectSort,38,330,20674,2027926,169618017
# StdSort,37,133,1449,18576,238349


plt.figure(figsize=(10, 5))
plt.title("Time Consumed - 100000 numbers")
plt.xlabel("Time used(log) - ns")
plt.ylabel("Alogrithm")
plt.xscale('log') 
plt.barh(result["sname"], result["100000_time"])
plt.show()
```

```python
plt.title("Time Consumed - 100 numbers")
plt.xlabel("Time used(log) - ns")
plt.ylabel("Alogrithm")
plt.xscale('log') 
plt.barh(result["sname"], result["100_time"])
plt.show()
```

```python
plt.title("Time Consumed - 1000 numbers")
plt.xlabel("Time used(log) - ns")
plt.ylabel("Alogrithm")
plt.xscale('log') 
plt.barh(result["sname"], result["1000_time"])
plt.show()
```

```python
def draw(alg_name):
    row = result.loc[result["sname"] == alg_name]
    print(row[["10_time", "100_time", "1000_time", "10000_time", "100000_time"]].values)
    plt.figure(figsize=(10, 5))
    plt.title("Time Consumed - " + alg_name)
    plt.xlabel("Number size")
    plt.ylabel("Time used(log) - ns")
    plt.xscale('log')
    plt.barh(["10_", "100_", "1000_", "10000_", "100000_"], row[["10_time", "100_time", "1000_time", "10000_time", "100000_time"]].values[0])
    # plt.legend()
    plt.show()
    
```

```python
draw("BubbleSort")
```

```python
draw("HeapSort")
```

```python
draw("HillSort")
```

```python
draw("InsertSort")
```

```python
draw("QuickSort")
```

```python
draw("SelectSort")
```
